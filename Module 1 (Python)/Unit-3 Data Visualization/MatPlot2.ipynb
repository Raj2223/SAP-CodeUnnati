```python
fig,ax=plt.subplots()

x=np.linspace(0,20,1000)
ax.plot(x,np.tan(x))
ax.axis('equal')

ax.annotate("local maximum",xy=(6,100),xytext=(100,500),arrowprops=dict(facecolor='black',shrink=0.05))
ax.annotate("local minimum",xy=(6*np.pi,-1),xytext=(-500,206),arrowprops=dict(arrowstyle="->",connectionstyle="angle3,angleA=0,angleB=-90"));
```

```python
import matplotlib.pyplot as plt
w=3
h=4
d=50
x=([3,3,1])
y=([3,3,1])
plt.figure(figsize=(w,h),dpi=d)
plt.axis([0,5,0,5])
label_x=1
label_y=3
arrow_x=2
arrow_y=4

arrow_properties=dict(facecolor='black',width=0.5,headwidth=4,shrink=0.1)

plt.annotate("maximum",xy=(arrow_x,arrow_y),xytext=(label_x,label_y),arrowprops=arrow_properties)
plt.plot(x,y)
```

```python
from mpl_toolkits import mplot3d

fig=plt.figure()

ax=plt.axes(projection='3d')

z=np.linspace(0,1,100)
x=z*np.sin(25*z)
y=z*np.cos(25*z)

ax.plot3D(x,y,z,'purple')
ax.set_title("3D Line Plot")
plt.show()
```
