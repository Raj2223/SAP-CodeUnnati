```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[1,2,3]
y=[2,4,3]
```

```python
plt.plot(x,y)
plt.show()

plt.bar(x,y)
plt.show()

plt.scatter(x,y)
plt.show()
```

```python
plt.plot(x,y,marker='*',color='r',linestyle='dotted')
plt.title("Data")
plt.xlabel("X-Lable")
plt.ylabel("Y-Lable")
plt.legend("Blue,green")
plt.show()
```
