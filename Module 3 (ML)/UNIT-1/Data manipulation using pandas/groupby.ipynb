## use groupby function to form groups based  on more than one category(i.e use more than one column to preform splitting)

```python
# creating the dataframe:

df = pd.read_csv('nba.csv')

# fitst grouping based on the team
# within each team we are grouping based on "position"

gkk = df.groupby(['Team', "Position"])
gkk2 = df.groupby(['Age', "Position"])
# printing the first value in each group 
gkk.first()
```

# Creating the pivot tables:

```python
df = pd.DataFrame({'A': ['John', "Boby", "Mina", "Peter", "Nicky"], "B":["Master", "Graduate", 'Graduate', "Master", 'Graduate'],
                  "C": [27,23,21,23,24]})
df
```

```python
# creating the pivot table and operationlise:
# simple piot table must have a dataframe:
# and an index/List of index:
table = pd.pivot_table(df, index = ["A", "B"])
table
```

```python
table2 = pd.pivot_table(df, values = 'A', index = ["B", 'C'], columns = ["B"], aggfunc = np.sum)
table2
```
