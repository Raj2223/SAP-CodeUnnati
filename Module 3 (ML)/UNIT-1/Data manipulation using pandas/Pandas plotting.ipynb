```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
ts = pd.Series(np.random.randn(1000), index = pd.date_range('1/1/2000', periods = 1000))
print(ts)
ts = ts.cumsum()
print(ts)
ts.plot()
plt.show()
```

```python

ts = pd.Series(np.random.randn(1000), index = pd.date_range('1/1/2000', periods = 1000))

df = pd.DataFrame(np.random.randn(1000, 4), index = ts.index, columns = list('ABCD'))
print(df)
df = df.cumsum()
plt.figure()
df.plot()
plt.show()
```
