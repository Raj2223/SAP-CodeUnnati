```python
car_data3 = {"Name":['Valiant', 'Duster 360', 'Merc 240D', 'Merc 320', 'Mer 280', 
                           'Merc 280C', 'Merc 450SE', 'Merc 450SL', 'Merc 450SLC', 
                           'Cadillac Fleetwood', 'Lincoln Continental',
                           'Crysler Imperial', 'Fiat 128', 'Honda Civik', 'Toyota Corolla'], 
             "Gear":[3,3,4,4,5,4,3,3,3,3,3,3,4,4,4], 
             "Cylinder": [6,8,4,4,6,6,8,8,8,8,8,8,4,4,4]}

# creating the daraframe for the data usign pandas dataframe function
car_df3 = pd.DataFrame(car_data3)

unique_Gear = pd.unique(car_df3.Gear)
unique_Cylinder = pd.unique(car_df3.Cylinder)

# printing the unique valures present in the gear column
print(f"Print the unique values present in the Gear column are: {unique_Gear}")

# printing the unique values present 
print(f"Unique values present in the Culinder column are : {unique_Cylinder}")

print("\n----------------------------------------------------------------------")
# now we count the values from the gear feature and and cylinder featur:

count_gear = car_df3["Gear"].value_counts()
count_cylinder = car_df3["Cylinder"].value_counts()
count_Name = car_df3["Name"].value_counts()
print(f"The counts of the gear:   \n{count_gear}")
print("\n")
print(f"The counts of the Cylinders:   \n{count_cylinder}")
print("\n")
print(f"The counts of the Car names:   \n{count_Name}")
print("\n\n------------------------------------------------------------------------")

print("Other method to count the values from the arributes:\n")
g_c = pd.value_counts(car_df3.Gear)
c_c = pd.value_counts(car_df3.Cylinder)
n_c = pd.value_counts(car_df3.Name)

print(f"The counts of the gear:   \n{g_c}")
print("\n")
print(f"The counts of the Cylinders:   \n{c_c}")
print("\n")
print(f"The counts of the car names:   \n{n_c}")
```

```python
# initializing the variables 0 for counting
name_unique = 0
gear_unique = 0
cylinder_unique = 0

# writing the seperate loo[ for each group

for i in n_c:
    name_unique+=1
for i in g_c:
    gear_unique+=1
for i in c_c:
    cylinder_unique+=1
print(f"The unique names: {name_unique}")
print(f"The unique gear: {gear_unique}")
print(f"The unique cylinder: {cylinder_unique}")
```
